### Converting Between Types

```python
# int()  : string, floating point, boolean  integer

print(int("5"))
print(int(5.6)) 
print(int(True))
print(int(False))
```

```python
# float() : string, integer , boolean floating point number

print(float(5))
print(float("5.5"))
print(float(True))
print(float(False))
```

```python
# str() : integer, float, boolean  string
display(str(25)) 
display(str(25.5))
display(str(True))
display(str(False))
```
