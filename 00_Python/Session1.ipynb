* a et b sont deux **identifiants** des variables définies
* Un identifiant doit etre significatif (nom , age, prénom..)
* Un nom de variable **doit** commencer par **une lettre** ou **_**
* Un nom de variable **ne peut pas** commencer par **un chiffre**
* Un nom de variable **ne peut contenir que** des caractères alphanumériques et des traits de soulignement (**A-z**, **0-9** et **_** )
* Les noms de variables sont **sensibles à la casse** (age, Age et AGE sont trois variables différentes)
* Un nom de variable **ne peut pas** etre un **keyword**

### Converting Between Types

```python
# int()  : string, floating point, boolean → integer

print(int("5"))
print(int(5.6)) 
print(int(True))
print(int(False))
```

```python
# float() : string, integer , boolean→ floating point number

print(float(5))
print(float("5.5"))
print(float(True))
print(float(False))
```

```python
# str() : integer, float, boolean → string
display(str(25)) 
display(str(25.5))
display(str(True))
display(str(False))
```

* ## Python Operators

   * Les opérateurs sont utilisés pour effectuer des opérations sur des variables et des valeurs.
   * Python divise les opérateurs en :
      * Opérateurs arithmétiques
      * Opérateurs d'affectation
      * Opérateurs de comparaison
      * Opérateurs logiques
      * Opérateurs d'identité
      * Opérateurs d'adhésion
      * Opérateurs au niveau du bit (Bitwise operators)

* **Les opérateurs arithmétiques** sont utilisés avec des valeurs numériques pour effectuer des opérations mathématiques

```python
50%3
```

```python
50//3
```

* **Les opérateurs d'affectation** sont utilisés pour affecter des valeurs aux variables

* **Les opérateurs de comparaison** sont utilisés pour comparer deux valeurs

* **Les opérateurs logiques** sont utilisés pour combiner des instructions conditionnelles

```python
5>2 and 5<8
```

* **Les opérateurs d'identité** sont utilisés pour comparer les objets, non pas s'ils sont égaux, mais s'ils sont en fait le même objet, avec le même emplacement mémoire

* **Les opérateurs d'appartenance** sont utilisés pour tester si une séquence est présentée dans un objet
