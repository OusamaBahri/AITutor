## Visualizing distributions of data :

```python
num_features = df.drop("Species",axis=1).columns
```

```python
# Plot distribution of sepal length
plt.hist(df["SepalLengthCm"],bins = 12)
plt.xlabel("sepal length cm")
plt.ylabel('number of flowers')
plt.title(f'Distribution of SepalLengthCm')
plt.show()
```

```python
# Plot distribution of sepal width
plt.hist(df["SepalWidthCm"], color = "skyblue")
plt.xlabel("sepal width cm")
plt.ylabel('number of flowers')
plt.title(f'Distribution of SepalWidthCm')
plt.show()
```

```python
# Plot distribution of petal length
plt.hist(df["PetalLengthCm"], color = "skyblue")
plt.xlabel("Petal length cm")
plt.ylabel('number of flowers')
plt.title(f'Distribution of PetalLengthCm')
plt.show()
```

```python
# Plot distribution of petal width
plt.hist(df["PetalWidthCm"], color = "skyblue")
plt.xlabel("Petal width cm")
plt.ylabel('number of flowers')
plt.title(f'Distribution of PetalWidthCm')
plt.show()
```

### Swarm plots :
It helps to represent the better representation of the distribution of values , here it  shows the distribution of data across several levels of one (or more) categorical variables(flower species in our case) such that those distributions can be compared.

```python
sns.set(style="whitegrid")
plt.figure(figsize=(12,10))
plt.subplot(2,2,1)
sns.swarmplot(x="Species",y="SepalLengthCm",data=df)
plt.subplot(2,2,2)
sns.swarmplot(x="Species",y="SepalWidthCm",data=df)
plt.subplot(2,2,3)
sns.swarmplot(x="Species",y="PetalLengthCm",data =df)
plt.subplot(2,2,4)
sns.swarmplot(x="Species",y="PetalWidthCm",data= df)
plt.show()
```

* The first plot reveals that there is a significant amount of overlap between the species on **sepal length** , so we cant't distinguish between our species based in this feature only .
* The second plot is the same as the first one , there is a significant overlap between the species on **Sepal width**.
* The third plot (**petal length**) seems promising since we can see that the Setosa species are well separated from Versicolor and Virginica, although there is some overlap between the Versicolor and Virginica, but not as bad as the the above two plots.
* The fourth plot(**petal width**)  looks good. There is slight intersection between the Setosa and Versicolor species, while the overlap between the Versicolor and Virginica is somewhat similar to that of petal length(Plot 3).


## Visualizing Relationships:

### Check the correlations between the features :
* **What is correlation ?**
    * Correlation explains how one or more variables are related to each other. The correlation coefficient ranges from -1 to 1.



```python
# Create the correlation matrix
corr = df[num_features].corr()
sns.heatmap(corr, annot=True)
plt.show()
```

--> we can see that :
* **Petal Width** and **Petal length** are highly positively correlated (0.97)
* **Petal width** and **Sepal length** are positively correlated (0.82)
* **Sepal Length** and **Petal length** are positively correlated (0.87)
* **Sepal width** and **Petal length** are negatively correlated (-0.42)
