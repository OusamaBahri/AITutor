* PassengerId: type should be integers
* Survived: Survived or Not
* Pclass: Class of Travel
* Name: Name of Passenger
* Sex: Gender
* Age: Age of Passengers
* SibSp: Number of Sibling/Spouse aboard
* Parch: Number of Parent/Child aboard
* Ticket: Ticket number
* Fare: Fare for ticket
* Cabin: Cabin Number
* Embarked: The port in which a passenger has embarked. C - Cherbourg, S - Southampton, Q = Queenstown

```python
import pandas as pd
import numpy as np
import seaborn as sns
import matplotlib.pyplot as plt
```

```python
#import data
data = pd.read_csv("titanic-passengers.csv",delimiter=";")
```

```python
data.describe()
```

```python
sns.countplot(x='Survived',hue="Sex" ,data=data,palette='pastel');
```

```python
plt.figure(figsize=(10,10))
sns.countplot(x='Survived', hue='Pclass', data=data, palette='pastel')
plt.title('Survived vs Pclass')
plt.show()
```

```python
corrMatrix = data.corr()
sns.heatmap(corrMatrix , annot=True)
plt.show()
```

```python
data.isnull().sum()
```

```python
data = data.drop("Cabin",axis=1)
```

```python
#fill with the mode
data.Embarked.fillna(data.Embarked.mode()[0],inplace=True)
```

```python
#fill with mean
data['Age'] = data['Age'].replace(np.nan, data.Age.mean())
```

```python
data.isnull().sum()
```

```python
 #replace male as 1 and female as 0
data["Sex"] = data["Sex"].map({"male": 1, "female":0})
```

```python
from sklearn.preprocessing import LabelEncoder
le = LabelEncoder()
data["Survived"] = le.fit_transform(data["Survived"].values)
```

```python
data["Embarked"] = le.fit_transform(data["Embarked"].values)
```

```python
# drop non usefull variables
data.drop(['PassengerId', 'Ticket'], axis=1, inplace=True)
```

```python
data.head()
```

```python
#Use the groupby function combined with the mean() to view the relation between Parch and survived 
# of parents / children aboard the Titanic(parch)
data.groupby(by=['Parch']).mean()["Survived"]
# every family consisted of 3 members 50 percent of them died
```

In fact, we dropped the column Names because we evaluated it as useless, well that's not the case, we can create a very useful feature from the names

your task now is to create a new column called Title which contain the appropriate Title for each individual ( hint : extract the title from the column Names)

## Visualize the correlation between Title and other features(e.g Sex, Fare, Age...)



```python
plt.figure(figsize=(15,8))
ax = sns.countplot(x='Title',hue = 'Sex' ,data=data,palette='pastel')
```

```python
sns.catplot(x="Title", y="Age", data=data , height=8.27,aspect=11.7/8.27)
```

As you should've seen, there is a lot of titles which makes the feature not very practical, let's try to group these titles into more useful ones

hint: use the following dictionary

```python
Title_Dictionary = {

                    "Capt":       "Officer",

                    "Col":        "Officer",

                    "Major":      "Officer",

                      "Dr":         "Officer",

                    "Rev":        "Officer" ,

                    "Jonkheer":   "Royalty",

                    "Don":        "Royalty",

                    "Sir" :       "Royalty",

                   "Lady" :      "Royalty" ,

                  "the Countess" : "Royalty",

                    "Dona":       "Royalty",

                    "Mme":        "Miss",

                    "Mlle":       "Miss",

                    "Miss" :      "Miss",

                    "Ms":         "Mrs",

                    "Mr" :        "Mrs",

                    "Mrs" :       "Mrs" ,

                    "Master" :    "Master"

                    }
```

```python
data.Title = data.Title.map(Title_Dictionary)
```

let's visualize the same correlations using these new titles now.

```python
sns.catplot(x="Title", y="Age", data=data , height=8.27, aspect=11.7/8.27)
```

```python
plt.figure(figsize=(15,8))
ax = sns.countplot(x='Title',hue = 'Sex' ,data=data,palette='pastel')
```

```python
sns.catplot(x="Title", y="Fare", data=data , height=8.27, aspect=11.7/8.27)
```

Finally, use the Parch and the SibSp columns to create a more useful feature, let's call it FamilySize

```python
data['family_size'] = data['SibSp'] + data['Parch'] + 1
print( data[["family_size","Survived"]].groupby(["family_size"], as_index = False).mean() )
```
